# Data Extraction

```python
import pandas as pd 
from glob import glob
import os
import time
```

```python
files = glob("../data/raw/*.csv")
print(len(files))
print(files)
```

```python
df_names = []
```

```python
for i in files :
    # import file 
    df = pd.read_csv(i)

    # add time and source
    df["extraction_timestamp"] = int(time.time())
    df["extraction_date"] = pd.to_datetime(df["extraction_timestamp"], unit="s")
    df["data_source"] = "Data Lake"
    
    #change name of df 
    file_name =os.path.basename(i)
    file_name = file_name.replace(".csv" , "")
    globals()[f"df_{file_name}"] = df

    # add to list
    df_names.append(f"df_{file_name}")

    #print new name
    print(f"Created: df_{file_name}")
```

```python
for i in df_names:
    
    df = globals()[i]
    length = df.shape
    nulls = df.isnull().sum().sum()
    
    print(f"{i}  ")
    print(f"   length : {length}")
    print(f"   nulls : {nulls}")
    print("-"*30)
```
